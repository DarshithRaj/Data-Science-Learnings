```python
# higly influence points
k = 2
n = len(df)
leverage_cutoff = 3*((k + 1)/n)
leverage_cutoff
```

```python
# points which are greater than leverage value treated as influencer observations
df[cooks>leverage_cutoff]
df.drop([70,76,78,79],inplace=True)
df.shape
```

```python
model = smf.ols('Price~Age_08_04+HP',data=df).fit()
model.summary()
```

```python
Y=df["Price"]
X=df[["Age_08_04","HP"]]
```

```python
# data partition
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test  = train_test_split(X,Y, test_size=0.2)
```

```python
from sklearn.linear_model import Lasso, Ridge
from sklearn.preprocessing import StandardScaler
```

```python
# Standardizing the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
coef_sm = model.params
print("Statsmodels coefficients:", coef_sm)
# Apply Lasso Regression
lasso = Lasso(alpha=1.0)
lasso.fit(X_train_scaled, Y_train)
print("Lasso coefficients:", lasso.coef_)
# Apply Ridge Regression
ridge = Ridge(alpha=1.0)
ridge.fit(X_train_scaled, Y_train)
print("Ridge coefficients:", ridge.coef_)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Create a Ridge regression model
ridge_model = Ridge(alpha=1.0)  # alpha is the regularization strength
ridge_model.fit(X, Y)

# Predicting using Ridge model
y_pred_ridge = ridge_model.predict(X)

# Evaluating the Ridge model performance
mse_ridge = mean_squared_error(Y, y_pred_ridge)
rmse_ridge = np.sqrt(mse_ridge)
r2_ridge = r2_score(Y, y_pred_ridge)

print("Ridge Regression:")
print("Root Mean Squared Error (RMSE):", np.round(rmse_ridge, 5))
print("R2 Score:", np.round(r2_ridge, 5))
```
